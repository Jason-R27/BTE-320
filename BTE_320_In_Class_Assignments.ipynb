```python
sales = float(input("Enter sales ($): "))
cogs = sales * 0.4
gp = sales - cogs

print("Sales: $", sales)
print("Cost of Goods Sold: $", cogs)
print("Gross Profit: $", gp)
```

```python
sales = float(input("Enter sales ($): "))
cogs = sales * 0.4
gp = sales - cogs
tr = float(input("Enter tax rate (%): "))
tax = tr * gp
net = gp - tax

print("Sales: $", sales)
print("Cost of Goods Sold: $", cogs)
print("Gross Profit: $", gp)
print ("Taxes: $", tax)
print("Net income: $:", net)
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operation (e.g., +, -, *, etc.): ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

```python
print("Hour","Fee")
for hour in range(1, 9):
  fee = 5 + 2.5 * hour
  print(f"{hour}    {fee}")
```

```python
print("Hour\tFee")
for hour in range(1, 9):
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10.0
  if fee > 20:
    fee = 20.0
  print(f"{hour}\t{fee}")
```

```python
print("Hour\tFee")
hour = 1
while hour < 9:
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10.0
  if fee > 20:
    fee = 20.0
  print(f"{hour}\t{fee}")
  hour += 1
```

Answer to Assignment 4

```python
import math

ticketPrice = 10
initAttendee = 20

print("Ad$\tProfit")
ad = 0
while ad < 201:
  revenue = (initAttendee + 2 * round(math.sqrt(ad))) * ticketPrice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{float(ad)}\t{float(profit)}")
  ad += 25
```

```python
import random

check = input("Do you want to play a game (y/n)? ").lower()

while check == 'y':
  player = input("Choose b/w rock, paper, scissors: ").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])
  print(f"Computer Move: ", computer)

  if player == computer:
    print("Draw")
  elif ((player == 'rock' and computer == 'scissors')
    or(player == 'scissors' and computer == 'paper')
    or(player == 'paper' and computer == 'rock')):
    print('Player wins!')
  else:
    print('Computer wins :(')
  check = input("Do you want to continue playing (y/n)? ")
```
