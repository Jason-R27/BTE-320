Assignment 14 - Greeting Class

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return "Instance of class 'Greeting'"

name = Greeting("Jason", "Rotenberg")
print(name)
name.display()
name.display(formal=True)
```

Assignment 15 - Student Class

```python
class Student:
  def __init__ (self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    self.__name = newName

  def __len__ (self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
      print(f"Enrolled in {newCourse}")
    else:
      print(f"Student {self.__name} is already enrolled in {newCourse}")
```

Assignemnt 16 - Classes Point and Rectangle

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle:
  def __init__ (self, p1, p2, p3, p4):
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4
    self.width = self.width_()
    self.height = self.height_()

  def width_(self):
    return self.p1.distance(self.p2)

  def height_(self):
    return self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height

p1 = Point(0,0)
p2 = Point(5,0)
p3 = Point(0,5)
p4 = Point(5,5)

z = Rectangle(p1, p2, p3,p4)
z.area()
```
