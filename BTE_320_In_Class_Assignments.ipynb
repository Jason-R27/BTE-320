Assignment 1

```python
sales = float(input("Enter sales ($): "))
cogs = sales * 0.4
gp = sales - cogs

print("Sales: $", sales)
print("Cost of Goods Sold: $", cogs)
print("Gross Profit: $", gp)
```

```python
sales = float(input("Enter sales ($): "))
cogs = sales * 0.4
gp = sales - cogs
tr = float(input("Enter tax rate (%): "))
tax = tr * gp
net = gp - tax

print("Sales: $", sales)
print("Cost of Goods Sold: $", cogs)
print("Gross Profit: $", gp)
print ("Taxes: $", tax)
print("Net income: $:", net)
```

```python
N1 = float(input("Enter number 1: "))
Sign = input("Enter Sign: ")
N2 = float(input("Enter number 2: "))

if Sign == "+":
  print(N1+N2)
if Sign == "-":
  print(N1-N2)
if Sign == "*":
  print(N1*N2)
if Sign == "/":
  print(N1/N2)
```
