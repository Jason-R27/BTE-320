Assignment 1

```python
sales = float(input("Enter sales ($): "))
cogs = sales * 0.4
gp = sales - cogs

print("Sales: $", sales)
print("Cost of Goods Sold: $", cogs)
print("Gross Profit: $", gp)
```

```python
sales = float(input("Enter sales ($): "))
cogs = sales * 0.4
gp = sales - cogs
tr = float(input("Enter tax rate (%): "))
tax = tr * gp
net = gp - tax

print("Sales: $", sales)
print("Cost of Goods Sold: $", cogs)
print("Gross Profit: $", gp)
print ("Taxes: $", tax)
print("Net income: $:", net)
```

```python
N1 = float(input("Enter number 1: "))
Sign = input("Enter Sign: ")
N2 = float(input("Enter number 2: "))

if Sign == "+":
  print(N1+N2)
if Sign == "-":
  print(N1-N2)
if Sign == "*":
  print(N1*N2)
if Sign == "/":
  print(N1/N2)
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operation (e.g., +, -, *, etc.): ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

Assignment 3

```python
print("Hour","Fee")
for hour in range(1, 9):
  fee = 5 + 2.5 * hour
  print(f"{hour}    {fee}")
```

```python
print("Hour\tFee")
for hour in range(1, 9):
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10.0
  if fee > 20:
    fee = 20.0
  print(f"{hour}\t{fee}")
```
